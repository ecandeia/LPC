## Filtros digitais



Qualquer sistema que modifica um sinal de uma forma predeterminada pode ser considerada como um filtro.  Filtros que modificam os sinais através da modificação dos seus conteúdos de frequência, ou seu espectro, são chamados filtros seletivos em frequência. 

Na disciplina Análise de Sinais e Sistemas vimos que os filtros analógicos mais comuns são os filtros passa baixa, passa alta, rejeita faixa e passa faixa, cujas as repostas em frequências, para o caso ideal, são ilustradas na Figura 1.



 Figura 1 - Filtros ideais analógicos
 
Por exemplo, o filtro passa-baixa ideal é aquele que deixa inalteradas as componentes de frequência de um sinal abaixo da frequência de corte especificada ($\omega_c$), e rejeita todos as componentes de frequência acima dessa frequência de corte. 

Entretanto, não podemos utilizar esses filtros ideais nos nossos experimentos no GNU Radio por dois motivos. Em primeiro lugar esses filtros são sistemas analógicos e o GNU Radio é um software digital e em segundo lugar esses filtros, conforme são ilustrados na Figura 1, são sistemas ideias e não causais e, portanto, possuem uma resposta ao impulso $h(t)$ de comprimento infinito.

Existem diversas soluções que podem ser adotadas para resolver esses problema.  A solução que adotaremos consiste em amostrar a resposta ao impulso do filtro ideal desejado e depois multiplicar por uma janela no domínio do tempo de tamanho N. Isso faz que a resposta ao impulso do filtro tenha um conjunto finito de N coeficientes. Esse procedimento permite obter filtros (com ordem N) com uma resposta ao impulso de duração finita. 


 Figura 2 - Multiplicação de um filtro ideal passa baixa por uma janela no domínio do tempo

Entretanto, os filtros obtidos ainda permanecem não causais, devido a $h(n)\neq0$ para $-(N-1)\leq n<0$. Para tornar o filtro causal é necessário realizar um deslocamento de $\alpha=\frac{N-1}{2}$ na resposta ao impulso truncada (Figura 3).


Figura 3 - Resposta ao impulso deslocada $h(n-\alpha)$, com $\alpha=20$

Observem que as aproximações de comprimento finito para a resposta de impulso ideal levam à presença de ondulações tanto na faixa de passagem como na faixa de rejeição do filtro, bem como a uma largura de transição diferente de zero entre a faixa de passagem e a faixa de parada (Figura 4).

Ambas as ondulações de banda passante / banda de rejeição e largura de transição são desvios indesejáveis, mas inevitáveis na resposta de um filtro ideal quando aproximado com uma resposta de impulso finita. 

Os filtros digitais obtidos através do procedimento aqui descrito são denomeninados filtros FIR (Finite Impulse Response), pois possuem uma resposta ao impulso com duração finita e são expressos por:

$$ y(n)=\sum_{k=0}^{N-1} h(k)x(n-k) $$

Para facilitar o projeto de um filtro FIR de ordem $N$ e frequência de corte $\Omega_c$ podemos calcular os $N$ termos da resposta ao impulso do filtro desejado utilizando as equações apresentadas a seguir, conforme for o tipo de filtro. Observe, que a frequência $\Omega$ é a frequência digital. Para fazer a conversão frequência analógica $\omega$  (rad/s) para digital $\Omega$(radianos) é necessário utilizar a seguinte equação $\Omega=\omega.T$, em que $T$ é o período de amostragem e $\omega_s=\frac{2 \pi}{T}$ é a frequência de amostragem.


**Passa Baixa**

$$h(n)=\begin{cases}\frac{\Omega_c}{\pi}, n=\alpha \\ \frac{sin\left[\Omega_c \left(n-\alpha\right)\right]}{\pi \left(n-\alpha\right)} , n\neq \alpha \end{cases} $$ 
 

**Passa Alta**  

$$h(n)=\begin{cases}
1-\frac{\Omega_c}{\pi}, n=\alpha, \\
\frac{sin\left[\pi \left(n-\alpha\right)\right]}{\pi \left(n-\alpha\right)} - \frac{sin\left[\Omega_c \left(n-\alpha\right)\right]}{\pi \left(n-\alpha\right)} , n\neq \alpha 
\end{cases}$$ 

**Passa Faixa** 

$$h(n)=\begin{cases}
\frac{\Omega_{c_2}-\Omega_{c_1}}{\pi}, n=\alpha, \\
\frac{sin(\Omega_{c_2} \left(n-\alpha\right))-sin(\Omega_{c_1} \left(n-\alpha\right))}{\pi n} , n\neq \alpha 
\end{cases}$$ 


**Rejeita Faixa**  

$$h(n)=\begin{cases}
1+\frac{ \Omega_{c_1}-\Omega_{c_2}}{\pi}, n=\alpha, \\
\frac{sin(\Omega_{c_1} \left(n-\alpha\right))-sin(\Omega_{c_2} \left(n-\alpha\right)) + sin(\pi\left(n-\alpha\right))}{\pi \left(n-\alpha\right)} , n\neq \alpha
\end{cases}$$ 



````python
# Carrega pacotes 

import matplotlib.pyplot as plt
import numpy as np
from numpy import cos, sin, log10, sqrt
from numpy.fft import fft, fftfreq, fftshift
from scipy.signal import convolve
import pandas as pd
from numpy.random import normal

pd.set_option("display.precision", 2)
figsize(10, 3)
````

````python

def FIR_passaBaixa(Ωc, N):
    """
    Ωc : Frequência de corte do filtro (rad)
    N  : Ordem do filtro 
    
    """
    α = (N-1)/2
    
    # alocação de variáveis
    
    h = np.zeros((N, 1)); #Resposta ao impulso do filtro, coeficientes do filtro

    # calcula dos coeficientes do filtro

    for n in range(0,N):
        if n == α:
            h[n] = Ωc/np.pi
        else:
            h[n] = sin(Ωc*(n-α))/(np.pi*(n-α))

    return h

def FIR_passaAlta(Ωc, N):
    """
    Ωc : Frequência de corte do filtro (rad)
    N  : Ordem do filtro 
    
    """
    α = (N-1)/2
    
    # alocação de variáveis
    
    h = np.zeros((N, 1)); #Resposta ao impulso do filtro, coeficientes do filtro

    # calcula dos coeficientes do filtro

    for n in range(0,N):
        if n == α:
            h[n] = 1 - (Ωc/np.pi)
        else:
            h[n] = ( sin(np.pi*(n-α)) - sin(Ωc*(n-α)) )/(np.pi*(n-α))

    return h

def FIR_passaFaixa(Ωc1, Ωc2, N):
    """
    Ωc1 : Frequência de corte 1 do filtro (rad)
    Ωc2 : Frequência de corte 2 do filtro (rad)
    N  : Ordem do filtro 
    
    """
    α = (N-1)/2
    
    # alocação de variáveis
    
    h = np.zeros((N, 1)); #Resposta ao impulso do filtro, coeficientes do filtro

    # calcula dos coeficientes do filtro

    for n in range(0,N):
        if n == α:
            h[n] = (Ωc2 - Ωc1)/np.pi
        else:
            h[n] = ( sin(Ωc2*(n-α)) - sin(Ωc1*(n-α)) )/(np.pi*(n-α))

    return h

def FIR_rejeitaFaixa(Ωc1, Ωc2, N):
    """
    Ωc : Frequência de corte do filtro (rad)
    N  : Ordem do filtro 
    
    """
    α = (N-1)/2
    
    # alocação de variáveis
    
    h = np.zeros((N, 1)); #Resposta ao impulso do filtro, coeficientes do filtro

    # calcula dos coeficientes do filtro

    for n in range(0,N):
        if n == α:
            h[n] = 1+ (Ωc1 - Ωc2)/np.pi
        else:
            h[n] = ( sin(Ωc1*(n-α)) - sin(Ωc2*(n-α)) + sin(np.pi*(n-α)))/(np.pi*(n-α))

    return h
````

### Projeto de um filtro FIR passa baixa

````python
# Resposta ao impulso do sistema

N = 50  # Ordem do filtro
fs = 8000 # frequência de amostragem em Hz
fc = 1000 # frequencia de corte do filtro em Hz

π=np.pi

ωs=2*π*fs
ωc=2*π*fc
T = 1/fs # Período de amostragem

# Conversão das frequências de amostragem e corte do filtro para  frequências digitais

Ωs = ωs*T
Ωc = ωc*T

n = range(0,N)


# Calculo da resposta em Frequencia de filtro
h = FIR_passaBaixa(Ωc, N)

# calculo da resposta em frequencia do filtro
Na = 1000 # número de amostras da FFT

H = fftshift(fft(h.flatten(), n=Na))
H = np.abs(H)
xf = fftshift(fftfreq(H.size, T))

#Plota gráficos
plt.figure()
(markers, stemlines, baseline) = plt.stem(n, h,'b',\
                                         use_line_collection=True,\
                                         label = 'coeficientes h(n)')

plt.setp(baseline, visible=False)
plt.setp(markers, marker='o', markersize=5, markeredgecolor="b", markeredgewidth=2)
plt.setp(stemlines, linestyle="--", color="b", linewidth=0.5 )


plt.title('Resposta ao impulso do filtro passa baixa')
plt.xlim(-1,N)
plt.xlabel('n')
plt.ylabel('h(n)')
plt.grid(color='k', linestyle='-', linewidth=0.5)


plt.figure()
plt.plot(xf, 20*log10(H),'b',label = 'Magnitude (dB)')

plt.xlim(0,xf.max())
plt.xlabel('f (Hz)')
plt.ylabel('| H(f) |')
plt.legend()
plt.grid(color='k', linestyle='-', linewidth=0.5)


#Gera tabela com os coeficientes h(n)
pd.set_option('display.max_columns', 500)
pd.options.display.float_format = '{:,.4f}'.format
pd.DataFrame(np.hstack((h)), columns=["h(n)"]).T
````

**Formata valores dos coeficientes para serem utilizados no bloco Decimating FIR Filter do GNU Radio:**

````python
print('h(n) = ', np.round(h.reshape(-1,), 5).tolist())
````

## Filtragem de um sinal

Filtros digitais são sistemas Lineares Invariantes ao Deslocamento. Portanto, a saída de uma operação de filtragem pode ser obitida através da operação de convolução entre o sinal de entrada e a resposta ao impulso do filtro. Essa operação é expressa pela seguinte equação:

$$ y(n)=x(n) \ast h(n) $$

Vamos agora analisar o efeito de um filtragem passa baixa sobre um sinal de onda quadrada. Esse sinal será gerado através da equação de síntese da série trigonométrica de Fourier, utilizando para tal as três principais componentes (conforme foi abordado no experimento 1). Um filtro passa baixa com frequência de corte de $f_c Hz$ é utilizado para filtrar as componentes do sinal de entrada que possuirem frequência fundamental maior que a frequência de corte do filtro. 

````python
# Geração de sinal do sinal de entrada

f0 = 125     # Frequência fundamental
fs = 8000  # Frequência de amostragem
T = 1/fs # Período de amostragem

t = np.arange(0, 100*40e-3, 1/fs)
π = np.pi
ω0=2*π*f0

x   = (4/π)*(sin(ω0*t) + sin(3*ω0*t)/3 + sin(5*ω0*t)/5) # onda quadrada com frequência fundamental f0 

# Projeto do filtro passa baixa

N  = 500 # Ordem do filtro
fc = 150  # frequencia de corte do filtro em Hz
ωc = 2*π*fc # frequencia de corte do filtro em rad/s
Ωc = ωc*T # frequencia de corte do filtro digital em rad

# Calculo da resposta em Frequencia de filtro
h = FIR_passaBaixa(Ωc, N)
th = np.arange(0,N*T, T)

# Calculo do sinal filtrado, através da operação de convloção entre h(n) e x(n)
h = h.flatten()
y = convolve(x,h, 'valid')
Nc = np.size(y)
ty = np.arange(0,Nc*T, T)
````

````python
tplot = 80e-3 # intervalo de tempo plotado

plt.figure()
plt.plot(t[t<=tplot], x[t<=tplot],'b')
plt.ylim(x.min(0)-0.1, x.max(0)+0.1)
plt.xlim(0,tplot)
plt.xlabel('tempo (s)')
plt.ylabel('x(t)')
plt.grid(color='k', linestyle='-', linewidth=0.5)

plt.figure()
plt.plot(th,h,'b')
plt.ylim(h.min(0)-0.01, h.max(0)+0.01)
plt.xlim(0,th.max(0))
plt.xlabel('tempo (s)')
plt.ylabel('h(t)')
plt.grid(color='k', linestyle='-', linewidth=0.5)

plt.figure()
plt.plot(ty[ty<=tplot], y[ty<=tplot],'b')
plt.ylim(y.min(0)-0.1, y.max(0)+0.1)
plt.xlim(0,tplot)
plt.xlabel('tempo (s)')
plt.ylabel('y(t)')
plt.grid(color='k', linestyle='-', linewidth=0.5)
````

Os efeitos da filtragem no domínio da frequência podem ser observados através da seguinte expressão:

$$Y(\Omega)= X(\Omega).H(\Omega).$$

Em que $X(\Omega)$ representa o espectro (Transformada de Fourier) do sinal de entrada, $H(\Omega)$ é a resposta em frequência do filtro e $Y(\Omega)$ é espectro do sinal de filtrado. Observem que o filtro atenua (praticamente elimina) as componentes de frequências acima da frequência de corte do filtro. Observem, também, que esse filtro não é ideal, quanto maior for a ordem do filtro mais semelhante ele será de um filtro ideal.

````python
from scipy import signal

plt.figure()
plt.magnitude_spectrum(x,Fs=fs,scale='dB',label='sinal', color='b');
plt.ylim(-100, 0)
plt.xlim(0,1000)
plt.xlabel('frequência (Hz)')
plt.ylabel('|X(ω)| (dB)')
plt.grid(color='k', linestyle='-', linewidth=0.5)

plt.figure()
w, Hf = signal.freqz(h, worN = 2000)
w = w/(2*np.pi)*fs
plt.plot(w, 20*log10(abs(Hf)), color='b')
plt.ylim(-100, 10)
plt.xlim(0,1000)
plt.xlabel('frequência (Hz)')
plt.ylabel('|H(ω)| (dB)')
plt.grid(color='k', linestyle='-', linewidth=0.5)

plt.figure()
plt.magnitude_spectrum(y,Fs=fs,scale='dB',label='sinal', color='b');
plt.ylim(-100, 0)
plt.xlim(0,1000)
plt.xlabel('frequência (Hz)')
plt.ylabel('|Y(ω)| (dB)')
plt.grid(color='k', linestyle='-', linewidth=0.5)
````

## Relação sinal-ruído


A relação sinal-ruído (ou razão sinal-ruído, *signal-to-noise ratio* ($\mathrm{SNR}$), em Inglês) é uma das grandezas mais importantes na engenharia de sistemas de comunicações. A $\mathrm{SNR}$ é um indicador da presença de ruído no sistema, ou seja, a presença de distorções aleatórias e indesejáveis que afetam os sinais que carregam informação, dificultando ou impossibilitando o processo de comunicação. 

O ruído de maior interesse prático é o ruído aditivo, que é adicionado ao sinal de informação $s(t)$ durante o processo de comunicação, como indicado na equação

$$y(t) = s(t) + n(t)$$

em que $n(t)$ representa o ruído e $y(t)$ o sinal ruidoso.

A $\mathrm{SNR}$ é definida como sendo a razão entre a potência de sinal $P_s$ e a potência do ruído $P_n$ observadas num dado sistema:

$$\mathrm{SNR} = \frac{P_s}{P_n}$$

em que $P_s = E\left[|s(t)|^2\right]$ e $P_n=E\left[|n(t)|^2\right]$, com $E[.]$ denotando valor esperado.

Quando expressa em decibéis (dB), a $\mathrm{SNR}$ é dada por

$$ \mathrm{SNR}_{dB} = 10\log_{10}P_s-10\log_{10}P_n.$$

Quanto maior a $\mathrm{SNR}$ maior a diferença entre a potência do sinal de interesse e a potência do ruído adicionado á mesma. Dessa forma, quanto maior a $\mathrm{SNR}$ melhor a qualidade do sinal.

Um dos modelos mais importantes para o ruído (talvez o modelo mais importante) é o modelo de ruído branco gaussiano aditivo (*additive white Gaussian noise* (AWGN), em Inglês). Nesse modelo, o ruído é representado por um processo aleatório gaussiano, ou seja, para cada instante $t$ no tempo, o ruído $n(t)$ adicionado ao sinal é dado por uma variável aleatória gaussiana de média zero e uma certa variância $\sigma^2$. 

````python
# ruído gaussiano branco
Namostras = 100000
σ2  = 0.5  # variância
μ   = 0    # média

σ        = sqrt(σ2) 
ruido    = normal(μ,σ, Namostras)  

# plotas as primeiras 1000 amostras
plt.plot(ruido[0:1000],linewidth = 0.8);
plt.xlim(0,1000)
plt.ylabel('amplitude')
plt.xlabel('amostra')
plt.grid()
````

Veja que o sinal ilustrado acima é a realização de um processo estocástico gaussiano estacionário. Cada amostra corresponde a uma realização de uma variável aleatória (v.a) gaussiana de média nula ($\mu=0$) e variância $\sigma^2$. As variáveis aleatórias, por sua vez, são independentes e idênticamente distribuídas (i.i.ds). Vamos calcular a potência do sinal acima, considerando que para um processo ergódico na correlação, podemos assumir:

$$P_x=E\left[X^2\right] = \mu^2 + \sigma^2\approx \frac{1}{N}\sum_{k=1}^{N} x^2[k]$$

Note que esta é a mesma expressão para cálculo da potência média de um sinal determinístico. Logo, temos que:

````python
# função para calcular a potência de um sinal
def potSinal(x):
    return (x**2).mean()

print('Potência do ruído = %.2f unidades de potência' %potSinal(ruido)) # veja a definição da função potSinal() acima
````

Perceba que a potência do ruído é numericamente igual à sua variância. De fato, o cálculo da potência de um sinal estocástico ergódico e com variáveis i.i.d. de média nula corresponde a uma estimativa da variância de cada v.a. que o compõe.

Vamos checar se o histograma da função amostra do ruído realmente corresponde à distribuição gaussiana solicitada:

````python
# fdp de uma variável aleatória gaussiana de média μ e desvio padrão σ
def gaussian(x, μ, σ):
    return np.exp(-((x - μ)**2)/(2*σ**2))/sqrt(2*np.pi*σ**2)
````

````python
plt.hist(ruido,51, density=True, label = 'histograma do ruído');
plt.grid()

x   = np.arange(-6*σ,6*σ, σ/10)
fdp = gaussian(x,μ,σ)

plt.plot(x,fdp,'k--',label ='fdp da distribuição gaussiana');
plt.legend();
plt.xlim(-3,3);
````

No domínio da frequência, $n(t)$ é caracterizado por sua densidade espectral de potência $S_N(f)$, assim como visto na figura a seguir

````python
fa = 8000   # frequência de amostragem considerada
plt.figure()
plt.psd(ruido,Fs=fa,label='densidade espectral do ruído',linewidth=0.8,NFFT=2048); # estima densidade espectral do sinal
plt.xlim(0,fa/2);
plt.grid(True);
plt.legend();

for line in plt.legend().get_lines():
    line.set_linewidth(2.0)
````

No gráfico acima, vê-se que $S_N(f)\approx -39$ dB/Hz. Perceba que a potência do ruído AWGN *distribui-se de maneira uniforme* sobre todo o espectro de componentes de frequência.

No sistema internacional de unidades (SI), $S_N(f)=N_0/2$ é dada em W/Hz, sendo $N_0$ uma constante representando a contribuição de cada componente de frequência para a potência média do ruído. Note que $S_N(f)$ é constante sobre todo o intervalo de frequências $[-\infty,\infty]$. Dessa forma, no caso ideal em que $t$ é uma variável contínua, a potência do ruído AWGN é infinita! Como nenhum sistema físico real possui banda infinita, um modelo mais adequado para análises práticas considera a presença de um ruído gaussiano limitado a uma banda $B$ que, por sua vez, pode ser obtido passando o ruído AWGN por um filtro passa-baixas ideal de banda $B$. Nesse caso, a potência média do ruído será dada por $P_N = N_0B$, pois

$$ P_N = \sigma^2 = \int_{-B}^{B} S_N(f)df $$

$$ \sigma^2 = \frac{N_0}{2}\int_{-B}^{B}df $$

$$ \sigma^2 = N_0B$$

Logo, 

$$ N_0 = \frac{\sigma^2}{B} $$

Uma simulação computacional, como a do GNU Radio, utiliza sinais discretos no tempo. Isto é, a variável temporal não é contínua e os sinais são representados à uma dada frequência de amostragem $f_a$. Como consequência do teorema de Nyquist, a frequência de amostragem $f_a$ impõe um intervalo limite $[-f_a/2,f_a/2]$ para a largura de banda máxima dos sinais discretos gerados na simulação. Dessa forma, sinais e sistemas representados numa simulação computacional naturalmente serão limitados em banda. Desse modo, ao gerar-se em simulação um processo aleatório AWGN de variância $\sigma^2$ à uma taxa de amostragem $f_a$, a densidade espectral de potência do ruído simulado será de $S_N(f) = 2\sigma^2/f_a$ W.

Vamos calcular o valor de $N_0$ esperado para o vetor de amostras de ruido gaussiano gerado anteriormente, para o qual consideramos $\sigma^2$ = 0.5 e $f_a$ = 8000 Hz:

````python
N0 = 2*σ2/fa

print('Valor de N0 em dB/Hz: %.2f dB/Hz' %(10*log10(N0)))
````

Note que o valor de $N_0$ calculado está de acordo com o valor estimado no gráfico da densidade espectral de potência obtido diretamente a partir das amostras do ruído geradas.

No que segue, vamos analisar o efeito na potência do ruído após passarmos o sinal por um filtro passa-baixas.

````python
# ruído gaussiano branco
Namostras = 100000
σ2  = 0.5  # variância
μ   = 0    # média

σ        = sqrt(σ2)
ruido    = normal(μ,σ, Namostras)  

# projeto do filtro passa baixa
N   = 1024    # Ordem do filtro
fc  = 2000    # frequencia de corte do filtro em Hz
T   = 1/fa    # período de amostragem em s
ωc  = 2*π*fc  # frequencia de corte do filtro em rad/s
Ωc  = ωc*T    # frequencia de corte do filtro digital em rad

# cálculo da resposta em frequência de filtro
h  = FIR_passaBaixa(Ωc, N)
th = np.arange(0,N*T, T)

# cálculo do ruído filtrado, através da operação de convolução entre h(n) e x(n)
h  = h.flatten()
ruidoFiltrado = convolve(ruido,h);

# plota espectro
plt.figure()
plt.psd(ruido,Fs=fa,label='densidade espectral do ruído',linewidth=1.5,NFFT=2048);
plt.psd(ruidoFiltrado,Fs=fa,label='densidade espectral do ruído filtrado',linewidth=1.5,NFFT=2048);
plt.xlim(0,fa/2);
plt.ylim(-150,)
plt.grid(True);
plt.legend();

for line in plt.legend().get_lines():
    line.set_linewidth(2.0)

# calcula potências antes e depois do filtro
print('Potência do ruído = %.2f unidades de potência' %potSinal(ruido))
print('Potência do ruído filtrado = %.2f unidades de potência' %potSinal(ruidoFiltrado))
````

Note que o filtro com banda de 2000 Hz filtrou a metade da banda do sinal original. Logo, metade dos componentes de frequência que compunham a forma de onda do ruído foram eliminados. Como a potência do ruído se distribui uniformemente sobre todos os seus componentes de frequência, a potência do sinal filtrado foi reduzida à metade da potência do sinal na entrada do filtro.

No exemplo a seguir, pode-se visualizar o efeito no domínio do tempo e no domínio da frequência da adição de ruído gaussiano a um tom senoidal.

````python
f0 = 500
fa = 8000
t = np.arange(0,10,1/fa)

# gera sinal senoidal
A = 2
y = A*sin(2*np.pi*f0*t)

# gera ruído gaussiano
σ2    = 0.25     # variância 
μ     = 0          # média 
σ     = sqrt(σ2)
ruido = normal(0,σ,len(t)) # gera realização do processo gaussiano

# plota sinais no domínio do tempo
plt.plot(t,ruido,'--',label ='ruido');
plt.plot(t,y+ruido,'--',label ='sinal+ruido');
plt.plot(t,y,'k--',label ='sinal');
plt.xlim(0,20e-3)
plt.ylim(-3,3)
plt.grid()
plt.legend()

# plota sinais no domínio da frequência
plt.figure()
plt.psd(y,Fs=fa,label='sinal', linewidth = 1.5,NFFT=2048);
plt.psd(y+ruido,Fs=fa,label='sinal+ruído', linewidth = 1.5,NFFT=2048);
plt.grid(True)
plt.legend()
plt.xlim(0,fa/2)
plt.ylim(-100,10);

for line in plt.legend().get_lines():
    line.set_linewidth(2.0)
````

````python
# relação sinal ruído em dB
SNR = 10*log10(potSinal(y)/potSinal(ruido))

print('SNR = %.2f dB' %SNR)
````

## Exemplo de aplicação com processamento de áudio


Sinais de áudio são representações eletrônicas de ondas sonoras - ondas mecânicas longitudinais que viajam pelo ar. Os sinais de áudio podem ser representados em formato digital ou analógico, e o seu o processamento pode ocorrer em qualquer um desses domínios. Os circuitos analógicos operam diretamente no sinal elétrico, enquanto os processadores digitais operam matematicamente a representação digital do sinal.

Os sinais de áudio possuem frequências aproximadamente na faixa entre 20 Hz e 20 kHz, que correspondem aos limites inferior e superior da audição humana. Os sinais de áudio podem ser sintetizados diretamente ou podem se originar em um transdutor, como um microfone. Por sua vez, alto-falantes convertem um sinal elétrico de áudio em som.

No que segue, vamos carregar e visualizar um sinal de áudio contido num arquivo .wav.


````python
import IPython
from scipy.io import wavfile
````

````python
# fa = 44100   # frequência de amostragem da placa de som

# carrega arquivo de áudio
fa, y = wavfile.read('audioTeste.wav')

t = np.arange(0,np.floor(len(y)/8)*1/fa,1/fa)
y = 2*y[:,1]/abs(y[:,1]).max(0)
y = y[0:int(np.floor(len(y)/8))]

tplot = 10
plt.figure()
plt.plot(t[t<=tplot], y[t<=tplot], linewidth=0.1, label = 'sinal de áudio');
plt.xlim(0,tplot)
plt.grid()
plt.xlabel('tempo (s)')
plt.ylabel('amplitude');

wavfile.write('sinal.wav', fa, y.astype(np.float32))

del y, t
````

No gráfico acima, os segundos iniciais do sinal de áudio carregado estão ilustrados. Perceba que tanto a amplitude quanto a frequência do sinal variam com o tempo. Frequências mais agudas (altas) geram tonalidades mais intensa de azul enquanto que frequências mais graves (baixas) geram uma tonalidades mais claras.

Toque o sinal de áudio carregado utilizando o player abaixo:

**Áudio original**

````python
IPython.display.Audio('sinal.wav')
````

Abaixo, pode-se tocar o sinal resultante após as seguintes operações:

* Adição de ruído
* Filtragem com filtro passa baixas
* Retificação do sinal (operação não-linear)


````python
## carrega arquivo de áudio
fa, y = wavfile.read('audioTeste.wav')
t = np.arange(0,np.floor(len(y)/8)*1/fa,1/fa)
y = 2*y[:,1]/abs(y[:,1]).max(0)
y = y[0:int(np.floor(len(y)/8))]

## Adição de ruído

# gera ruído gaussiano
σ2    = 0.0025    # variância 
μ     = 0         # média 
σ     = sqrt(σ2)

ruido = normal(0,σ,len(y)) # gera realização do processo gaussiano

plt.figure()
plt.psd(y[t<=tplot] + ruido[t<=tplot], Fs=fa, label ='sinal ruidoso', linewidth=1.5, NFFT=2048);
plt.psd(y[t<=tplot], Fs=fa, label ='sinal', linewidth=1.5, NFFT=2048);
plt.grid()

wavfile.write('sinalRuidoso.wav', fa, (y+ruido).astype(np.float32))       # gera áudio ruidoso

## Distorção não-linear

wavfile.write('sinalDistNaoLinear.wav', fa, abs(y).astype(np.float32))    # gera áudio com uma distorção não-linear (y=|x|)
plt.psd(abs(y[t<=tplot]), Fs=fa, label ='sinal retificado',linewidth=1.5,NFFT=2048);

## Filtragem passa-baixas

# Projeto do filtro passa baixa
N   = 1024    # Ordem do filtro
fc  = 1000    # frequencia de corte do filtro em Hz
T   = 1/fa    # período de amostragem em s
ωc  = 2*π*fc  # frequencia de corte do filtro em rad/s
Ωc  = ωc*T    # frequencia de corte do filtro digital em rad

# Calculo da resposta em Frequencia de filtro
h  = FIR_passaBaixa(Ωc, N)
th = np.arange(0,N*T, T)

# Calculo do sinal filtrado, através da operação de convolução entre h(n) e x(n)
h  = h.flatten()
y_ = convolve(y,h)
Nc = np.size(y_)
ty = np.arange(0,Nc*T, T)

wavfile.write('sinalFiltrado.wav', fa, y_.astype(np.float32))

plt.psd(y_[ty<=tplot], Fs=fa, label ='sinal filtrado',linewidth=1.5,NFFT=2048);
plt.legend();
plt.xlim(0, fa/2);
plt.ylim(-140,);

for line in plt.legend().get_lines():
    line.set_linewidth(2.0)
    
````

````python
# apaga variáveis para salvar memória
del y_, ty, ruido, t, y, ruidoFiltrado
````
